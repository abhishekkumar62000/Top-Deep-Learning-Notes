

# Multi-Layer Perceptron (MLP)

## MLP kya hota hai?

Multi-Layer Perceptron ek **Artificial Neural Network (ANN)** hai jisme:

* Input layer
* One or more hidden layers
* Output layer

hote hain.

Single layer perceptron sirf **linear problems** solve karta tha.
MLP **complex non-linear problems** solve karta hai.

Example:

* Image classification
* Fraud detection
* Recommendation systems
* Medical prediction

---

## MLP Structure

Basic structure:

```
Input Layer → Hidden Layer(s) → Output Layer
```

Example:

```
x1  x2  x3
 |   |   |
 ---------
 Hidden Layer
 ---------
 Hidden Layer
 ---------
 Output
```

Deep Learning tab hota hai jab **hidden layers zyada ho jayein**.

---

# Neural Network ka Flow (Important)

MLP me 2 main steps hote hain:

### 1. Forward Propagation

Data input se output tak jata hai.

Process:
Input → Multiply weights → Add bias → Activation → Next layer

### 2. Backpropagation

Error calculate hota hai aur weights update hote hain.

Goal:
Loss minimize karna.

---

# ANN Notation (Bahut Important for AI Engineers)

Ye notation research papers aur interviews me use hota hai.

Main components:

### Input

x = input features

Example:

```
x = [x1, x2, x3]
```

---

### Weights

Har connection ka ek weight hota hai.

Notation:

```
W
```

Layer wise:

```
W¹  → Input to Hidden
W²  → Hidden to Output
```

---

### Bias

Neuron ko shift karta hai.

Notation:

```
b
```

Layer wise:

```
b¹
b²
```

---

# Neural Network Equation (Layer wise)

MLP me equation is tarah hoti hai:

Hidden layer output:

```
z¹ = W¹x + b¹
a¹ = activation(z¹)
```

Output layer:

```
z² = W²a¹ + b²
a² = activation(z²)
```

Yaha:

* z = weighted sum
* a = activation output

---

# Full ANN Notation (General Form)

Agar network me L layers hain:

```
z[l] = W[l] a[l-1] + b[l]
a[l] = g(z[l])
```

Yaha:

l = layer number
g = activation function

# Python Implementation (MLP Example)

Using sklearn:

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

# dataset
X, y = make_moons(n_samples=500, noise=0.2)

X_train, X_test, y_train, y_test = train_test_split(X, y)

# MLP model
model = MLPClassifier(
    hidden_layer_sizes=(10,5),
    activation='relu',
    max_iter=1000
)

model.fit(X_train, y_train)

accuracy = model.score(X_test, y_test)
print("Accuracy:", accuracy)
```

# 1️⃣ Neural Network Architecture

Input → Hidden Layer → Hidden Layer → Output

Example:

```
Input (4 features)
      ↓
Hidden Layer 1 (8 neurons, ReLU)
      ↓
Hidden Layer 2 (4 neurons, Tanh)
      ↓
Output Layer (1 neuron, Sigmoid)
```

---

# 2️⃣ Forward Propagation Formula

Layer output:

[
Z = W.X + b
]

Z = WX + b

Activation:

[
A = f(Z)
]

---

# 3️⃣ Activation Functions (Implementation)

```python
import numpy as np

# sigmoid
def sigmoid(z):
    return 1/(1+np.exp(-z))

# tanh
def tanh(z):
    return np.tanh(z)

# relu
def relu(z):
    return np.maximum(0,z)

# prelu
def prelu(z, alpha=0.01):
    return np.where(z>0, z, alpha*z)

# softmax
def softmax(z):
    exp = np.exp(z - np.max(z))
    return exp / exp.sum(axis=1, keepdims=True)
```

---

# 4️⃣ Activation Derivatives (Backpropagation)

```python
def sigmoid_derivative(a):
    return a*(1-a)

def tanh_derivative(a):
    return 1 - a**2

def relu_derivative(z):
    return (z > 0).astype(float)
```

---

# 5️⃣ Dataset Load

```python
import pandas as pd

df = pd.read_csv("student_placement.csv")

X = df.drop("placed", axis=1).values
y = df["placed"].values.reshape(-1,1)
```

---

# 6️⃣ Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

---

# 7️⃣ Neural Network Initialization

Network structure:

```
4 input
8 hidden
4 hidden
1 output
```

```python
np.random.seed(42)

input_size = 4
hidden1 = 8
hidden2 = 4
output_size = 1

# weights
W1 = np.random.randn(input_size, hidden1)
b1 = np.zeros((1, hidden1))

W2 = np.random.randn(hidden1, hidden2)
b2 = np.zeros((1, hidden2))

W3 = np.random.randn(hidden2, output_size)
b3 = np.zeros((1, output_size))
```

---

# 8️⃣ Forward Propagation Implementation

```python
def forward(X):

    global Z1,A1,Z2,A2,Z3,A3

    Z1 = np.dot(X, W1) + b1
    A1 = relu(Z1)

    Z2 = np.dot(A1, W2) + b2
    A2 = tanh(Z2)

    Z3 = np.dot(A2, W3) + b3
    A3 = sigmoid(Z3)

    return A3
```

---

# 9️⃣ Loss Function (Binary Cross Entropy)

```python
def compute_loss(y, y_pred):

    m = len(y)

    loss = -(1/m) * np.sum(
        y*np.log(y_pred) + (1-y)*np.log(1-y_pred)
    )

    return loss
```

---

# 🔟 Backpropagation Implementation

```python
def backward(X, y, y_pred, lr=0.01):

    global W1,W2,W3,b1,b2,b3

    m = len(X)

    # output layer gradient
    dZ3 = y_pred - y
    dW3 = (1/m) * np.dot(A2.T, dZ3)
    db3 = (1/m) * np.sum(dZ3, axis=0, keepdims=True)

    # hidden layer 2
    dA2 = np.dot(dZ3, W3.T)
    dZ2 = dA2 * tanh_derivative(A2)

    dW2 = (1/m) * np.dot(A1.T, dZ2)
    db2 = (1/m) * np.sum(dZ2, axis=0, keepdims=True)

    # hidden layer 1
    dA1 = np.dot(dZ2, W2.T)
    dZ1 = dA1 * relu_derivative(Z1)

    dW1 = (1/m) * np.dot(X.T, dZ1)
    db1 = (1/m) * np.sum(dZ1, axis=0, keepdims=True)

    # update weights
    W3 -= lr * dW3
    b3 -= lr * db3

    W2 -= lr * dW2
    b2 -= lr * db2

    W1 -= lr * dW1
    b1 -= lr * db1
```

---

# 1️⃣1️⃣ Training Loop

```python
epochs = 1000

for i in range(epochs):

    y_pred = forward(X)

    loss = compute_loss(y, y_pred)

    backward(X,y,y_pred)

    if i%100==0:
        print("Epoch:",i,"Loss:",loss)
```

---

# 1️⃣2️⃣ Prediction

```python
def predict(X):

    pred = forward(X)

    return (pred > 0.5).astype(int)
```

---

# 1️⃣3️⃣ Accuracy

```python
from sklearn.metrics import accuracy_score

y_pred = predict(X)

print("Accuracy:", accuracy_score(y, y_pred))
```

# 1️⃣5️⃣ PyTorch Implementation (Industry Level)

Real world me scratch rarely likhte hain.

```python
import torch
import torch.nn as nn

model = nn.Sequential(

    nn.Linear(4,8),
    nn.ReLU(),

    nn.Linear(8,4),
    nn.Tanh(),

    nn.Linear(4,1),
    nn.Sigmoid()
)

loss_fn = nn.BCELoss()

optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

Training:

```python
for epoch in range(100):

    pred = model(torch.tensor(X).float())

    loss = loss_fn(pred, torch.tensor(y).float())

    optimizer.zero_grad()

    loss.backward()

    optimizer.step()

    print(epoch, loss.item())
```
